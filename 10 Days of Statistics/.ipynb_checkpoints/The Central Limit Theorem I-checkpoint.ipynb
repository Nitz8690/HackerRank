## Day 6: The Central Limit Theorem I

```python
# Enter your code here. Read input from STDIN. Print output to STDOUT

import math

def cdf(x, m, sd):
    return 0.5 * (1 + math.erf((x - m)/(sd * math.sqrt(2))))
    

max_weight = 9800
n = 49
m = 205
sd = 15

print("%.4f" % round(cdf(max_weight, n * m, math.sqrt(n) * sd), 4))
```
